```python


# Identify the rows where the location column contains the continent name
continent_rows = df['location'].isin(['Asia', 'Europe', 'Africa', 'North America', 'South America', 'Oceania','World'])

# Create a new column for continent names and populate it

df.loc[continent_rows & df['continent'].isnull(), 'continent'] = df.loc[continent_rows & df['continent'].isnull(), 'location']

df.loc[df.location == 'Africa']
```

```python
for i, location in enumerate(df['location']):
    if location == 'European Union':
        df.loc[i, 'continent'] = 'Europe'
df.loc[df['location'] == 'European Union']
```

```python
df.continent.fillna('Class',inplace= True)
```

```python
df.loc[df.continent == 'Class']
```

```python
df.to_csv('owid-covid-data-clean.csv')
```
